```python
import pandas as pd
import os
```

```python
input_dir = "/Volumes/WD/CloudStation/lung_cancer_10X/шоишо║/2019.05.10"
cluster_xlsx = "annotation_results_by_cluster.xlsx"
km_luad = "TCGA_km_stage_LUAD.xlsx"
km_lusc = "TCGA_km_stage_LUSC.xlsx"
```

```python
def read_data(cell_name: str):
    u"""
    read data from xlsx
    :param cell_name: str
    """
    cluster = pd.read_excel(os.path.join(input_dir, f"{cell_name}/{cluster_xlsx}"))
    luad = pd.read_excel(os.path.join(input_dir, f"{cell_name}/{km_luad}"))
    lusc = pd.read_excel(os.path.join(input_dir, f"{cell_name}/{km_lusc}"))
    return cluster, luad, lusc
```

```python
basal, basal_luad, basal_lusc = read_data("Basal")
```
